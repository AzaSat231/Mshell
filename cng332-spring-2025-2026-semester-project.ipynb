#### 1. Interactive execution

- Your *MShell* shall read a *resource file* named `.mshellrc`&mdash;note the preceding dot character in the file name&mdash;that will be placed at the top level of your development directory, and it shall contain *environment variable* settings of the format

  ```Bash
  VARIABLE=VALUE
  ```
  <br>

  Spaces around the `=` should be considered immaterial, unlike in *Bash*.

  Note that environment variables may or may <u>not</u> have enclosing `'...'` or `"..."` around them. For example,
  
  ```Bash
  filepath1=/home/user/Documents
  filepath2='/home/user/Documents'
  filepath3="/home/user/Documents"
  ```
  <br>

  are essentially *equivalent*.

- Your *MShell* shall provide a *default* prompt, which shall be `mshell> `

    - The prompt string of your *MShell* shall be customizable by setting the `PS1` *environment variable*. The string that will be assigned to `PS1` shall be used as is without any variable substitution.
    
    - The default value of the *MShell* prompt may be *hardcoded*
    
    - When, for example, the user enters
    
      ```Bash
          mshell> PS1='metu> '
      ```
      <br>

      <br>
      <br>
      
      ```Bash
          metu>  
      ```
      <br>
      

- Your *MShell* shall only *exit* when the user provides the *built-in* `exit` command or presses `[Ctrl + D]` at the MShell prompt

    - Your shell need <u>not</u> respond to any other *signal* situations except this one!

#### 2. Built-in commands

You shall implement the following built-ins&mdash;the following is a subset of what *Bash* provides:

| Command | Description |
| --- | --- |
| `cd` | Change directory |
| `pwd` | Print working directory |
| `exit` | Terminate the shell |
| `echo` | Print text |
| `export` | Set environment variables |
| `unset` | Remove environment variables |
| `env` | Display environment |
| `history` | Show command history within a given shell session |
| `kill` | Send signals |

Consult your GNU/Linux system and/or your WSL installation about the *functionality* of these built-in commands.

Keep in mind that you are only required to provide the *functionality* described below.

##### `cd [dir]`

The `cd` command should change the *current working directory*, if specified with a correct path.

If no path is specified, then the the *current working directory* must be the *home directory* of the current user, as specified in the `$HOME` environment variable within *mshell*.

##### `pwd`

The `pwd` command shall print out the *absolute path* of the current working directory to the screen.

##### `exit`

The `exit` command shall cause *mshell* to *exit* to the calling context.

##### `echo` "...."

The `echo` command shall print simple messages in strings to the screen. 

The `echo` command shall also support *variable substitution*. For example:

```Bash
x=5
echo "The x-value is ${x}"
```

##### `export`

The `export` command shall turn on the *exportability* property of a given *environment variable*. Normally variables are defined using the `VARIABLE=VALUE` syntax. However, *exportable* variable shall be defined using the

```Bash
export VARIABLE=VALUE
```

assignment syntax.

*Exportable* variables are those that will also be valid in *mshells* created from another *mshell* instance!

##### `unset`

````python
The `unset` command removes the definition of a given *environment variable*.
````

##### `env`

The `env` command displays all the *environment variable* definitions (variable names and values) to the screen.

##### `history`

The `history` command shall display an *index* value and the complete command as typed by the user as in Bash. For example:

```Bash
2003  history
2004  ls
2005  ls -1
2006  ls -al
2007  cat websites.txt | wc
2008  history
```

These index values should be referrable to be able to *repeat* the execution of those commands without having to rewrite them. 

When an `!` symbol is used before a command index, your shell shall repeat that command.

For example, if the user wished to repeat the `cat websites.txt | wc` command, he/she can type the command

```Bash
mshell> !2007
```

where `mshell>` is the *prompt*. After typing `!2007`, the shell automatically re-runs `cat websites.txt | wc`.

##### `kill`

The `kill` command shall only support

```Bash
kill -<SIGNAL_NUMBER>
```

where `<SIGNAL_NUMBER>` is an *integer* signal number one can obtain by typing

```Bash
kill -l
```

in a Bash terminal window.

#### 3. External command execution

- Your *MShell* shall be able to execute any external executable with that commands's command-line arguments as long as that executable is on the *path* defined within your *MShell* environment.

- The *path* shall be defined as in Bash or other similar shells, that is, using the `PATH` *environment variable*. 

- An initial value of `PATH` may be hardcoded in your code. However, a customized setting may be placed in the `.mshellrc` resource file, and, if that is the case, your *MShell* shall use that *path* setting, not the hardcoded one. 

- The `PATH` specification format shell be identical to that in Bash:
  > See https://www.gnu.org/software/bash/manual/bash.html#index-PATH

- Your *MShell* shall look up the command entered by the user in the directories specified in the `PATH` environment variable, unless the user already specified the *absolute path* of the external command/program.

#### 4. Pipelines

Your *MShell* shall support arbitrary number of *pipes* entered as one does in a Bash shell using the `|` character:

```Bash
mshell> cmd_1 | cmd_2
mshell> cmd_1 | cmd_2 | cmd_3
mshell> cmd_1 | cmd_2 | cmd_3 | ...
```

That is, the *output* of the command to *left* of a pipe shall be *redirected* to the *input* of the command to the *right* of the pipe, and so on.

#### 5. I/O Redirection

Your *MShell* shall support *input* and *output* redirection for *stdin*, *stdout*, and *stderr* as in *Bash*. Specifically, your shell shall support the following forms of I/O redirection:

```Bash
mshell> cmd_1 > cmd_2
mshell> cmd_1 < cmd_2 > cmd_3
mshell> cmd_1 1> <output_1> 2> output_2
mshell> cmd_1 < input_1 > output_1 2>&1 
```

#### 6. Environment variables

- Your shell shall allow *environment variables* to be defined and referred to.

- An *environment variable* shall be defined using an assignment statement:
    
    ```Bash
    VARIABLE=VALUE
    ```
    <br>

    Bash does <u>not</u> allow spaces around the `=` sign. However, your implementation may relax this requirement.

- A given *environment variable* shall be accessed by value via either the `$VARIABLE` or `${VARIABLE}` syntax, as in *Bash*.

#### 7. Signal handling

Your *MShell* shall support signaling only via the `kill` command.

As mentioned, your shell executable shall <u>not</u> *exit* on a `SIGINT` (2) signal. How your shell implementation shall exit has already been defined above. Please refer to that discussion.

#### 8. Background jobs

Your *MShell* shall allow the user to execute external commands *in the background* if the user places a `&` symbol at the end of his/her command.

Running in the background means that the user shall get the *MShell* prompt back as soon the provided command is executed without waiting for that program to end. For example,

```Bash
mshell> python program.py &
```

shall cause your *MShell* to execute the indicated program in the background so that it can present the user with the `mshell> ` prompt right away, as in *Bash*.
